```python
import cv2
import os
import matplotlib.pyplot as plt

folder = '/home/sandra/Escritorio/Thyroid segmentation'

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
    
        if img is not None:
            images.append(img)
        plt.imshow(img)
    matrix_size = img.shape
    return  (images, matrix_size)
```

```python
images , matrix = load_images_from_folder(folder)
print ( matrix[0])
```
